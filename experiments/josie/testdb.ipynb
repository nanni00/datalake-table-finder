```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import Row
```

```python
from tools.josiedataprep.preparation_functions import _create_token_set
```

```python
def prepare_tuple(t, mode):
    return [t[1], _create_token_set(t[0][1], mode)]
```

```python
def create_index(mode, output_integer_set_file, output_inverted_list_file):
    spark = SparkSession \
        .builder \
        .appName("CreateJosieDataWithMongoDB") \
        .master('local')\
        .config("spark.mongodb.input.uri", "mongodb://127.0.0.1:27017/optitab.wikitables") \
        .config("spark.mongodb.output.uri", "mongodb://127.0.0.1:27017/optitab.wikitables") \
        .config('spark.jars.packages', 'org.mongodb.spark:mongo-spark-connector_2.12:3.0.1') \
        .getOrCreate()

    table_collection = spark.read\
        .format('mongo')\
        .option("uri", "mongodb://127.0.0.1:27017/optitab.wikitables") \
        .load() \
        .select('_id', 'content') \
        .rdd \
        .map(list)
    
    return table_collection.zipWithUniqueId().map(lambda t: prepare_tuple(t, mode)).top(10) # saveAsTextFile(output_integer_set_file)
    
    
    #return table_collection
```

```python
coll = create_index('set', 'integer_sets', 'inverted_lists')
```
