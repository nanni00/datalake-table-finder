```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
from code.utils.settings import DefaultPath
```

```python
stat1 = pd.read_csv(DefaultPath.data_path.wikitables + 'global_stat_1.csv')
stat2 = pd.read_csv(DefaultPath.data_path.wikitables + 'global_stat.csv')
```

```python
stat1
```

```python
stat2
```

```python
stat2['reading time (%)'] /= 100
stat2['rebuilding time (%)'] /= 100
stat2['embedding time (%)'] /= 100
stat2['storing time (%)'] /= 100
```

### Time Analyses

```python
plt.yscale('log')
plt.plot(stat1['n_tables'], stat1['total time (s)'], 'o--')
plt.plot(stat2['n_tables'], stat2['total time (s)'], 'o--')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
for title, ax, data in zip(['without labeling', ' with labeling'], axes, [stat1, stat2]):
    ax.set_xscale('log')
    ax.set_ylim(0, 1)
    ax.plot(data['n_tables'], data['reading time (%)'],     'o-', label='reading time (%)')
    ax.plot(data['n_tables'], data['rebuilding time (%)'],  'o-', label='rebuilding time (%)')
    ax.plot(data['n_tables'], data['embedding time (%)'],   'o-', label='embedding time (%)')
    ax.plot(data['n_tables'], data['storing time (%)'],     'o-', label='storing time (%)')
    ax.legend()
    ax.set_xlabel('n_tables')
    ax.set_ylabel('time (%)')
    ax.set_title(title)
```

### Disk Space Analyses 

```python
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(15, 5))
ax0.set_xscale('log')
ax0.plot(stat1['n_tables'], stat1['db size (GB)'],      'o--', label='db size (GB)')
ax0.plot(stat1['n_tables'], stat1['index size (GB)'],   'o--', label='index size (GB)')
ax0.set_xlabel('n_tables')
ax0.set_ylabel('size (GB)')
ax0.set_title('DB and index sizes')
ax0.legend()

ax1.set_xscale('log')
ax1.plot(stat1['n_tables'], stat1['db-over-index frac'], 'ro-')
ax1.set_xlabel('n_tables')
ax1.set_ylabel('db-over-index size fraction')
ax1.set_title('db-over-index size fraction')
```
